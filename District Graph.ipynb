```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def VisualiseGraph(G):
    LabelDict = {}
    Nodesize = []
    Colours = []
    for node in list(G.nodes()):
        LabelDict[node] = G.nodes[node]['name']
        if G.nodes[node]['type'] == 'crime':
            Nodesize.append(200)
            Colours.append('red')
        if G.nodes[node]['type'] == 'human class':
            Nodesize.append(400)
            Colours.append('yellow')
        if G.nodes[node]['type'] == 'district':
            Nodesize.append(100)
            Colours.append('green')
        if G.nodes[node]['type'] == 'place':
            Nodesize.append(600)
            Colours.append('blue')
        
    nx.draw(G,labels = LabelDict,node_size=Nodesize,node_color = Colours)
    plt.show()
```

```python
CrimeDF = pd.read_csv('crimeAM.csv')
columns = CrimeDF.columns
crimes = list(columns[1:])
```

```python
VisualiseGraph(Graph)
```

```python
def PredictCrime(PC,District,G):
    PCCrimes,DistrictCrimes,PlaceCrimes = [],[],[]
    for node in G.neighbors(NodeIDList[PC]):
        if G.get_edge_data(NodeIDList[PC],node)['weight']>0:
            PCCrimes.append(G.nodes[node]['name'])
    for node in G.neighbors(NodeIDList[District]):
        if G.get_edge_data(NodeIDList[District],node)['weight']>0:
            DistrictCrimes.append(G.nodes[node]["name"])
    Crimes = list(set(PCCrimes).union(set(DistrictCrimes)))
    CrimeAverageProbability = []
    for c in Crimes:
        p1 = G.get_edge_data(NodeIDList[c],NodeIDList[PC])['weight']
        p2 = G.get_edge_data(NodeIDList[c],NodeIDList[District])['weight']
        p = (p1+p2)/2
        CrimeAverageProbability.append((c,p))
        CrimeAverageProbability = sorted(CrimeAverageProbability,key=lambda x:x[1],reverse=True)
        
    #for node in G.neighbors(NodeIDList[Place]):
     #   PlaceCrimes.append(G.nodes[node]['name'])
    return CrimeAverageProbability
        
    
```
